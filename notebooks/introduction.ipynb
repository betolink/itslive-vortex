```python
points=[(-47.1, 70.1),
        (-46.1, 71.2)]

velocities = itslive.velocity_cubes.get_time_series(points=points, variables=["v","vx", "vy"])
```

```python
import matplotlib as mpl
from matplotlib import pyplot as plt

fig, ax = plt.subplots()


color_index = 0
for vel in velocities:
    latitude = velocities[0]["requested_point_geographic_coordinates"][1]
    longitude = velocities[0]["requested_point_geographic_coordinates"][0]
    
    point_label = f"Lon: {longitude}, Lat: {latitude}"
    velocities[0]["time_series"].v.plot(ax=ax, linestyle="None", marker="o", markersize=2, c=plt.cm.tab10(color_index), label=point_label)
    color_index+=1
```
